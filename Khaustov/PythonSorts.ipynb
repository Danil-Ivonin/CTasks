```python
import random
import time
```

```python
arr_len = 30_000
```

```python
def bubblesort(array):
    if len(array) <= 1:
        return array
    else:
        iterations = len(array) - 1
        for i in range(iterations):
            for j in range(iterations-i):
                if array[j] > array[j+1]:
                    array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
def quicksort(array):
   if len(array) <= 1:
       return array
   else:
       q = random.choice(array)
   l_array = [n for n in array if n < q]
 
   e_array = [q] * array.count(q)
   b_array = [n for n in array if n > q]
   return quicksort(l_array) + e_array + quicksort(b_array)
```

```python
def generate_array(amount = 10_000):
    arr = [i for i in range(amount)]
    random.shuffle(arr)
    return arr
```

```python
def merge(arr, l, m, r):
	n1 = m - l + 1
	n2 = r - m

	L = [0] * (n1)
	R = [0] * (n2)

	for i in range(0, n1):
		L[i] = arr[l + i]

	for j in range(0, n2):
		R[j] = arr[m + 1 + j]

	i = 0
	j = 0
	k = l

	while i < n1 and j < n2:
		if L[i] <= R[j]:
			arr[k] = L[i]
			i += 1
		else:
			arr[k] = R[j]
			j += 1
		k += 1

	while i < n1:
		arr[k] = L[i]
		i += 1
		k += 1

	while j < n2:
		arr[k] = R[j]
		j += 1
		k += 1
```

```python
def mergeSort(arr, l, r):
	if l < r:

		# Same as (l+r)//2, but avoids overflow for
		# large l and h
		m = l+(r-l)//2

		# Sort first and second halves
		mergeSort(arr, l, m)
		mergeSort(arr, m+1, r)
		merge(arr, l, m, r)
```

```python
arr = generate_array(arr_len)
print(arr)
bubble_start = time.time()
bubblesort(arr)
bubble_end = time.time()
print(arr)
```

```python
arr = generate_array(arr_len)
print(arr)
quick_start = time.time()
quicksort(arr)
quick_end = time.time()
print(arr)
```

```python
arr = generate_array(arr_len)
print(arr)
merge_start = time.time()
mergeSort(arr, 0, arr_len-1)
merge_end = time.time()
print(arr)
```

```python
print("for array len = \t" + str(arr_len))
print("bubblesort time = \t" + str(bubble_end - bubble_start))
print("quicksort time = \t" + str(quick_end - quick_start))
print("mergesort time = \t" + str(merge_end - merge_start))
```
