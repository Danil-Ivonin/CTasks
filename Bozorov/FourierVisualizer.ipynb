```python
import matplotlib.pyplot as plt
from matplotlib import cbook, cm
from matplotlib.colors import LightSource
import numpy as np
```

```python
# Читаем файл
with open('input.txt', 'r') as file:
    input_data = file.read().split()
tmp = []
for num in input_data:
    tmp.append(float(num))

# Преобразуем в numpy-массив
input = np.array(tmp)
```

```python
# Читаем файл
with open('output.txt', 'r') as file:
    output_data = file.read().split()
tmp = []
for num in output_data:
    tmp.append(float(num))

# Преобразуем в numpy-массив
output = np.array(tmp)
```

```python
plt.plot(input)
```

```python
plt.plot(output)
```

```python
from scipy.fft import fft

real = fft(input)
plt.plot(real)
```

```python
import random
# Создаем координатную сетку
x = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        x[i][j] = random.randrange(0, 1000) / 1000

# Создаем фигуру и 3D-ось
fig = plt.figure()
ax1 = fig.add_subplot()
ax2 = fig.add_subplot()

# Визуализация поверхности
ax1.imshow(x, cmap=cm.Reds)

# Отображаем график
plt.show()
```
